```python
import pandas as pd
import numpy as np
import plotly
import cufflinks as cf
import plotly.express as px
# %matplotlib inline
```

```python
df_sep = pd.read_csv("C:/Users/YX/Downloads/Fed Fund Rate Projection.csv")
df_sep['Report Date'] = pd.to_datetime(df_sep['Report Date'])
df_sep["Projected Year"] = pd.Categorical(df_sep["Projected Year"])
df_sep.sort_values(by=['Report Date','Projected Year', 'Percent'], ascending=[True, True, True], inplace=True)
df_sep['Report Date'] = df_sep['Report Date'].astype(str)
df_sep.reset_index(drop=True)
```

```python
fig = px.scatter(df_sep, x="Projected Year", y="Percent", animation_frame="Report Date",
                 size="Participants",size_max=40,
                 range_y=[-.1,5], height=1100,
                 color="Participants")

fig.update_yaxes(nticks=15)
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1500

fig.show()
```

```python
# Download interative dot plot as html file
plotly.offline.plot(fig, filename="C:/Users/YX/Downloads/Fed Fund Rate Projection.html")
```
